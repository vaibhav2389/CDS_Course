```python
# Step 1 : Independent and dependent variable creation

X =df[['TV','Radio','Newspaper']]  # X variable contains all inputs 
y = data.Sales # y is output variable here. 

df['TV'].count()
df['Radio'].count()
df['Newspaper'].count()
```

```python
y
```

```python
# Step 2 : Testing and Training the data. 
from sklearn.model_selection import train_test_split
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2,test_size=0.2)

# Random_state is seed setter in order to get exact values everytime you run this . Exact same to what we saw in random function. 
```

```python
X_train.shape
y_train.shape
X_test.shape
y_train.shape
# X_train.head()
```

```python
# Step3 : Creating the Machine Learning Model. 

from sklearn.linear_model import LinearRegression

LR = LinearRegression()  # object creation.

LR.fit(X_train,y_train) #training the linear regression model. 



y_predict=LR.predict(X_test)  # Prediction Output for test data. 
```

```python
y_pre_train = LR.predict(X_train)
```

```python
y_test.head()
```

```python
y_pre_train
```

```python
# Step 4 : Model Evaluation. 

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
#test_score 

r2score = r2_score(y_test,y_predict)   # Between Test Data and Prediction of Test Data by Model. 
r2score
```

```python
r2score_2 = r2_score(y_train, y_pre_train)  # Output between Training output and model predicted training model
```

```python
# Getting all other Model evaluation numbers here. 


# MEAN SQUARED ERROR
import math
MSE = mean_squared_error(y_test,y_predict)
MSE
```

```python
# ROOT MEAN SQUARED ERROR

RMSE = math.sqrt(mean_squared_error(y_test,y_predict))
RMSE
```

```python
# MEAN ABSOLUTE ERROR

MAE = mean_absolute_error(y_test,y_predict)
MAE
```
